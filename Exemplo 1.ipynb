## Desenvolvimento do Código

```python
# Importação das Bilbiotecas:

from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Definição do sistema de EDOs:

def EDO(t,y):
    
    CA = y[0]
    CB = y[1]
    CC = y[2]
    T = y[3]
    
    dCA_dt = -k0*np.exp(-Ea/(R*T))*CA
    dCB_dt = 2*k0*np.exp(-Ea/(R*T))*CA
    dCC_dt = k0*np.exp(-Ea/(R*T))*CA
    dT_dt = k0*np.exp(-Ea/(R*T))*CA*(-dH)/(ro*Cp)
    
    return [dCA_dt, dCB_dt, dCC_dt, dT_dt]
```

```python
# Dados/Equações auxiliares

Ea = 48500
R = 8.314
k0 = 82000000
dH = -72800
Cp = 1750
ro = 880
```

```python
# Parâmetros para resolução do sistema de EDOs

t_interval = [0, 40]
y0 = [100, 0, 0, 300]

resp = solve_ivp(EDO, t_interval, y0, t_eval = np.linspace(0, 40, 100))
```

```python
# Gráfico dos resultados

plt.plot(resp.t, resp.y[0], label = 'CA')
plt.plot(resp.t, resp.y[1], label = 'CB')
plt.plot(resp.t, resp.y[2], label = 'CC')
plt.xlabel('Tempo')
plt.ylabel('Concentração')
plt.legend()
plt.show()
```

```python
# Gráfico dos resultados

plt.plot(resp.t, resp.y[3], label = 'T')
plt.xlabel('Tempo')
plt.ylabel('Temperatura')
plt.legend()
plt.show()
```
